# Bayes Theorem Activity

Let's use Bayes Theorem to estimate the probability a penguin is one of the three species based on their bill measurment. 

- Choose one of the features (bill_length_mm for example).
- Choose one of the species.
- Fix a cut-off value for the freature below or above which you will guess your penguin is from the species.

So for our purposes:  A = penguin is our species; B = penguins feature is above (or below) our cut-off. 

Estimate P(A|B).

In the process you will need to estimate P(A) and P(not A). You have choices on how to do this: 

1. You could do some research and see if you can learn what the proportion of penguins is that is your species.

2. You could assume that your dataset is a random sample and that the proportions of the penguins in it from each species represents their proportion in the wild. 

Neither of these options is great. 1. I am not really sure what we know about the relative sizes of the populations of these penguins. 2. The evidence is, based on the data about islands etc, that our sample is not completely random and therefore it is not clear it is representative. But, as is usual when working with real data, the best we can do is proceed with the data we have, but be clear about the assumptions and what they mean for our result.
