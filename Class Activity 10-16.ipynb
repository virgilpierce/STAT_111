To answer this we are going to do the following resampling experiment:

We will bootstrap the sample (sample it with replacement to rebuild it to the same size), which will introduce some randomness. Then divide it into two sets, one the size of the number of ckd patients in the original and the other the size of the number of notckd patients in the original and then ask what the difference in the means or medians of the two sets is.
